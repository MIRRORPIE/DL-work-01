```
# write a function to plot some vectors from list

def plot_vectors(vecs):
  colors=["r","g","b","y","i","o"]
  i=0
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units="xy",angles="xy",scale=1,color=colors[i%len(colors)])
    i+=1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

## Vector Dot Products

```
import matplotlib.pyplot as plt
import numpy as np

# write a function to plot some vectors from list

def plot_vectors(vecs):
  colors=["r","g","b","y","i","o"]
  i=0
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units="xy",angles="xy",scale=1,color=colors[i%len(colors)])
    i+=1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
vecs_arr = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-2]),np.asarray([0,0,4,-1])]
plot_vectors(vecs_arr)
```

```
# i and j forms
a=np.asarray([3,4])
b=np.asarray([4,-1])
```

* \vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}|\cos(\theta) = a_x b_x + a_y b_y

$\vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}|\cos(\theta) = a_x b_x + a_y b_y$

```
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```
#visualise the dot product graphically

```

* a_b = |\vec{a}| \cos(\theta) = |\vec{a}| \frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}

$a_b = |\vec{a}| \cos(\theta) = |\vec{a}| \frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```
a_b=np.dot(a,b) / np.linalg.norm(b)
print(a_b)
```

unit vector along any vector

*\vec{a_b} = a_b\frac{\vec{b}}{|\vec{b}|}

$\vec{a_b} = a_b\hat{b} = a_b\frac{\vec{b}}{|\vec{b}|}$

```
# so unit projection af vector a on b in unit terms

vec_unit_a_b = (a_b / np.linalg.norm(b)) * b
print(vec_unit_a_b)
```

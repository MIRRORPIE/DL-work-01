```
def sigmoid(x, w, b):
  return(1 / (1 + np.exp(-(w*x + b))))
```

## Plotting Loss for a given dataset

```
# finding the Y for the given X....

import numpy as np
import matplotlib.pyplot as plt

w_unknown = 0.5
b_unknown = 0.25

def sigmoid(x, w, b):
  return(1 / (1 + np.exp(-(w*x + b))))
np.random.seed(0)            # it's concept is very very importtant
X = np.random.random(25) * 20 - 10   #25 random numbers in range -10 to 10
print("X is------------------->",X)
Y = sigmoid(X, w_unknown, b_unknown)
print("Y is------------------->",Y)
```

```
plt.plot(X, Y, "*")  #scatter plot
plt.show()
```

By bruteforce method here we are finding the w and b,    </br> if X and Y are given...

```
#calculate loss

def calculate_loss(X, Y, w_est, b_est):    #w_estimate and b_estimate
  loss = 0
  for x, y in zip(X, Y):
    loss += (y - sigmoid(x, w_est, b_est))**2
  return(loss)
```

```
"""we are creating the W and B meshgrid. between -1 and 1 , nearly 100 values inbetween"""

W = np.linspace(-1, 1, 101)
B = np.linspace(-1, 1, 101)

WW, BB = np.meshgrid(W, B)

Loss = np.zeros(WW.shape)
```

```
count = 0
for i in range(WW.shape[0]):
  for j in range(WW.shape[1]):
    count += 1
    Loss[i, j] = calculate_loss(X, Y, WW[i, j], BB[i, j])
    
print(count)
```

## Class for Sigmoid Neuron

```
print(np.random.randn(3,6))    #randomly created array
```

```
print(np.random.random(6))     #we cannot create array by this
```

```
np.random.seed(0)
fgb = np.random.random(1)
print(fgb, type(fgb))
```

```
# we have the X and Y and we have to find the best W and B values

class SigmoidNeuron:
  
  def __init__(self):
    self.w = None
    self.b = None
    
  def perceptron(self, x):                # this is basically the z = wx+b
    return (np.dot(x,self.w.T) + self.b)
  
  def sigmoid(self, z):
    return (1.0 / (1.0 + np.exp(-z)))
  
  def grad_w(self, x, y):
    y_pred = self.sigmoid(self.perceptron(x))
    return ((y_pred - y) * y_pred * (1 - y_pred) * x)
  
  def grad_b(self, x, y):
    y_pred = self.sigmoid(self.perceptron(x))
    return ((y_pred - y) * y_pred * (1 - y_pred))
  
  def fit(self, X, Y, epochs=1, learning_rate=1, initialise=True):
    
    #initialise the w and b
    if initialise:
      np.random.seed(0)
      self.w = np.random.randn(1, X.shape[1])
      np.random.seed(0)
      self.b = np.random.random(1)
    
    for i in range(epochs):
      dw = 0
      db = 0
      for x, y in zip(X, Y):
        dw += self.grad_w(x, y)
        db += self.grad_b(x, y)
        
      self.w -= learning_rate * dw
      self.b -= learning_rate * db
      
```

## Fit for toy data

```
X = np.array([[2.5, 2.5], [4, -1], [1, -4], [3, 1.25], [2, 4], [1, 5]])
Y = np.array([1, 1, 1, 0, 0, 0])
print("X shape is---->",X.shape)

sn = SigmoidNeuron()
sn.fit(X, Y, 1, 0.25, True)

print(sn.w, sn.b)

for i in range(10000):
  sn.fit(X, Y, 1, 0.25, False)
  
print(sn.w, sn.b)
```

```
def plot_sn(X, Y, sn):
  X1 = np.linspace(-10,10,101)
  X2 = np.linspace(-10,10,101)
  XX1, XX2 = np.meshgrid(X1, X2)
  YY = np.zeros(XX1.shape)
  for i in range(X2.size):
    for j in range(X1.size):
      val = np.asarray([X1[j], X2[i]])
      YY[i, j] = sn.sigmoid(sn.perceptron(val))
  
  plt.contourf(XX1, XX2, YY, cmap=my_cmap, alpha=0.6)    # my_cmap defined above in the previous cells
  plt.scatter(X[:,0], X[:,1], c=Y, cmap=my_cmap)
  plt.plot()
  
```

```
def plot_sn(X, Y, sn, ax):
  X1 = np.linspace(-10,10,100)
  X2 = np.linspace(-10,10,100)
  XX1, XX2 = np.meshgrid(X1, X2)
  YY = np.zeros(XX1.shape)
  for i in range(X2.size):
    for j in range(X1.size):
      val = np.asarray([X1[j], X2[i]])
      YY[i, j] = sn.sigmoid(sn.perceptron(val))
  
  ax.contourf(XX1, XX2, YY, cmap=my_cmap, alpha=0.6)
  ax.scatter(X[:,0], X[:,1], c=Y, cmap=my_cmap)
  ax.plot()
  
```

```
# we have the X and Y and we have to find the best W and B values

class SigmoidNeuron:
  
  def __init__(self):
    self.w = None
    self.b = None
    
  def perceptron(self, x):                # this is basically the z = wx+b
    return (np.dot(x,self.w.T) + self.b)
  
  def sigmoid(self, z):
    return (1.0 / (1.0 + np.exp(-z)))
  
  def grad_w(self, x, y):
    y_pred = self.sigmoid(self.perceptron(x))
    return ((y_pred - y) * y_pred * (1 - y_pred) * x)
  
  def grad_b(self, x, y):
    y_pred = self.sigmoid(self.perceptron(x))
    return ((y_pred - y) * y_pred * (1 - y_pred))
  
  def fit(self, X, Y, epochs=1, learning_rate=1, initialise=True):
    
    #initialise the w and b
    if initialise:
      np.random.seed(0)
      self.w = np.random.randn(1, X.shape[1])
      np.random.seed(0)
      self.b = np.random.random(1)
    
    for i in range(epochs):
      dw = 0
      db = 0
      for x, y in zip(X, Y):
        dw += self.grad_w(x, y)
        db += self.grad_b(x, y)
        
      self.w -= learning_rate * dw
      self.b -= learning_rate * db
      
  def predict(self, X):
    Y_pred = []
    for x in X:
      y_pred = self.sigmoid(self.perceptron(x))
      Y_pred.append(y_pred)
    return np.array(Y_pred)
      
```

## Loss plotting

```
from sklearn.metrics import mean_squared_error
```

```
class SigmoidNeuron:
  
  def __init__(self):
    self.w = None
    self.b = None
    
  def perceptron(self, x):                # this is basically the z = wx+b
    return (np.dot(x,self.w.T) + self.b)
  
  def sigmoid(self, z):
    return (1.0 / (1.0 + np.exp(-z)))
  
  def grad_w(self, x, y):
    y_pred = self.sigmoid(self.perceptron(x))
    return ((y_pred - y) * y_pred * (1 - y_pred) * x)
  
  def grad_b(self, x, y):
    y_pred = self.sigmoid(self.perceptron(x))
    return ((y_pred - y) * y_pred * (1 - y_pred))
  
  def fit(self, X, Y, epochs=1, learning_rate=1, initialise=True, display_loss=False):
    
    #initialise the w and b
    if initialise:
      np.random.seed(0)
      self.w = np.random.randn(1, X.shape[1])
      np.random.seed(0)
      self.b = np.random.random(1)
      
    if display_loss:
      loss = {}
      
    
    for i in range(epochs):
      dw = 0
      db = 0
      for x, y in zip(X, Y):
        dw += self.grad_w(x, y)
        db += self.grad_b(x, y)
        
      self.w -= learning_rate * dw
      self.b -= learning_rate * db
      
      if display_loss:
        Y_pred = self.sigmoid(self.perceptron(X))
        loss[i] = mean_squared_error(Y_pred, Y)

        
      if display_loss:
        plt.plot(loss.values())
        plt.xlabel('Epochs------>')
        plt.ylabel('Mean Squared Error-------->')
        
  def predict(self, X):
    Y_pred = []
    for x in X:
      y_pred = self.sigmoid(self.perceptron(x))
      Y_pred.append(y_pred)
    return np.array(Y_pred)
```
